### Load the data

```python
df = pd.read_csv("D:\\JOB ASSIGNMENT\\p1\\Data & Resources\\Data\\Customer_Data.csv")
df.head(5)
#skim(df)
```

```python
df.info()
```

```python
# GENDER PERCENTAGE %

Total_genders = df.groupby("Gender")["Gender"].count()

Total_count_of_genders = Total_genders.sum()

percentage_ratio = (Total_genders/Total_count_of_genders)*100

gender_stats = Total_genders.to_frame(name="Total Genders")
gender_stats["Percentage"] = percentage_ratio

percentage_genders = gender_stats.reset_index()
percentage_genders
```

```python
# Contract PERCENTAGE %

Total_Contract = df.groupby("Contract")["Contract"].count()

Total_count_of_Contract = Total_Contract.sum()

percentage_ratio = (Total_Contract/Total_count_of_Contract)*100

Contract_stats = Total_Contract.to_frame(name="Total Contract")
Contract_stats["Percentage"] = percentage_ratio

percentage_Contract = Contract_stats.reset_index()
percentage_Contract
```

```python
# State PERCENTAGE %

Total_State = df.groupby("State")["State"].count()

Total_count_of_State = Total_State.sum()

percentage_ratio = (Total_State/Total_count_of_State)*100

State_stats = Total_State.to_frame(name="Total count of State")
State_stats["Percentage"] = percentage_ratio

percentage_State = State_stats.reset_index()
percentage_State = State_stats.sort_values(by="Total count of State", ascending= False)
percentage_State
```

```python
# Customer Status PERCENTAGE %

grouped = df.groupby('Customer_Status').agg(
    TotalCount=('Customer_Status', 'count'),
    TotalRev=('Total_Revenue', 'sum')
).reset_index()

# Step 3: Calculate the total revenue for all statuses
total_revenue = df['Total_Revenue'].sum()

# Step 4: Calculate the revenue percentage for each status
grouped['RevPercentage'] = (grouped['TotalRev'] / total_revenue) * 100
grouped
```
